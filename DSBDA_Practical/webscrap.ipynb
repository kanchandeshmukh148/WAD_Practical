```python
!pip install requests beautifulsoup4 pandas
import requests

from bs4 import BeautifulSoup
import pandas as pd

life_quotes = []  # will store tuples of (quote, author)
page_count = 1

while True:
    url = f"http://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(url)

    if res.status_code != 200:
        break

    soup = BeautifulSoup(res.text, "html.parser")
    quotes = soup.find_all("div", class_="quote")

    if not quotes:
        break

    for q in quotes:
        text = q.find("span", class_="text").get_text(strip=True)
        author = q.find("small", class_="author").get_text(strip=True)
        life_quotes.append((text, author))

    page_count += 1

# Create DataFrame
df = pd.DataFrame(life_quotes, columns=["quote", "author"])
print(df)

# Optional: Save to CSV
df.to_csv("quotes.csv", index=False)
print("Quotes saved to quotes.csv")
```

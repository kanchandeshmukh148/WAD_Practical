```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\Kanchan\Desktop\Projects\DSBDA\Facebook .csv", sep=";")
df
```

```python
# 1. Total Interactions Histogram
df['Total Interactions'].hist()
plt.title("Total Interactions")
plt.show()
```

```python
# 2. Type vs Interactions
df.groupby('Type')['Total Interactions'].mean().plot(kind='bar')
plt.title("Type vs Interactions")
plt.show()
```

```python
# 3. Paid vs Interactions
df.groupby('Paid')['Total Interactions'].mean().plot(kind='bar')
plt.title("Paid vs Interactions")
plt.show()
```

```python
df.groupby('Post Month')['Total Interactions'].mean().plot(kind='bar')
plt.title("Month vs Interactions")
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.countplot(x='Type', data=df)

plt.title("Count of Post Types")

plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.boxplot(
    x='Type',
    y='Total Interactions',
    data=df
)

plt.title("Type vs Total Interactions Boxplot")

plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(
    x='Lifetime Post Total Reach',
    y='Total Interactions',
    hue='Type',
    data=df
)

plt.title("Reach vs Total Interactions")

plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.lineplot(
    x='Post Month',
    y='Total Interactions',
    data=df
)

plt.title("Post Month vs Total Interactions")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

correlation = df.corr(numeric_only=True)

sns.heatmap(correlation, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```
